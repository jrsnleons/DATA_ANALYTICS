## Data Exploration

**TASK: Create a correlation heatmap.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ucimlrepo import fetch_ucirepo

# Fetch the Parkinsons dataset using fetch_ucirepo
parkinsons = fetch_ucirepo(id=174)

# Extract the features and targets
X = parkinsons.data.features
y = parkinsons.data.targets

# Combine features and targets into a single DataFrame
data = pd.concat([X, y], axis=1)

# Compute the correlation matrix
correlation_matrix = data.corr()

# Create the heatmap
plt.figure(figsize=(15, 12))
heatmap = sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap - Parkinsons Dataset')
plt.show()
```

**TASK: What are the top 5 correlated frequencies with the target\label?**

*Note: You many need to map the label to 0s and 1s.*

*Additional Note: We're looking for **absolute** correlation values.*

```python
# Ensure the target labels are mapped to 0s and 1s if needed
# y = y.apply(lambda x: 1 if x == 'yes' else 0)  # Example mapping, adjust as needed

# Combine features and targets into a single DataFrame
data = pd.concat([X, y], axis=1)

# Compute the correlation matrix
correlation_matrix = data.corr()

# Get absolute correlation values of features with the target
target_correlations = correlation_matrix.iloc[-1][:-1].abs()

# Find the top 5 features with the highest absolute correlation with the target
top_5_features = target_correlations.nlargest(5)

print("Top 5 features most correlated with the target:")
print(top_5_features)
```

## Train | Test Split

Our approach here will be one of using Cross Validation on 90% of the dataset, and then judging our results on a final test set of 10% to evaluate our model.

**TASK: Split the data into features and labels, and then split into a training set and test set, with 90% for Cross-Validation training, and 10% for a final test set.**

*Note: Do not forget to put a random_state for reproducibility.*

```python
from sklearn.model_selection import train_test_split

# Combine features and targets into a single DataFrame (if not already done)
data = pd.concat([X, y], axis=1)

# Split the data into features (X) and labels (y)
X = data.drop(columns=['status'])  # Assuming 'status' is the target column
y = data['status']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(f'Training set size: {X_train.shape[0]} samples')
print(f'Test set size: {X_test.shape[0]} samples')
```

**TASK: Create a PipeLine that contains both a StandardScaler and a KNN model**

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Define the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('knn', KNeighborsClassifier())
])

# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# Print the pipeline steps to confirm
print(pipeline)
```

**TASK: Perform a grid-search with the pipeline to test various values of k and report back the best performing parameters.**

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for k values to test
param_grid = {'knn__n_neighbors': range(1, 31)}  # Testing k values from 1 to 30

# Set up the GridSearchCV with the pipeline and parameter grid
grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy')

# Perform the grid search on the training data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best parameters: {best_params}")
print(f"Best cross-validation accuracy: {best_score:.4f}")
```

**(HARD) TASK: Using the .cv_results_ dictionary, see if you can create a plot of the mean test scores per K value.**

```python
import matplotlib.pyplot as plt

# Extract mean test scores and corresponding k values from the .cv_results_ dictionary
mean_test_scores = grid_search.cv_results_['mean_test_score']
k_values = grid_search.cv_results_['param_knn__n_neighbors'].data

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, mean_test_scores, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Mean Test Score')
plt.title('Mean Test Scores for Different k Values')
plt.grid(True)
plt.xticks(range(1, 31))
plt.show()
```

### Final Model Evaluation

**TASK: Using the grid classifier object from the previous step, get a final performance classification report and confusion matrix.**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Predict on the test set using the best estimator from the grid search
y_pred = grid_search.best_estimator_.predict(X_test)

# Calculate accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print(f'Test set accuracy: {test_accuracy:.4f}')

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)

# Classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:')
print(class_report)
```
